```python
dataset = dataset_original.copy()
```

```python
# TODO: What's the average review length?

dataset['text'].str.len().mean()

# no preprocessing, including spaces and punctuation
```

```python
# TODO: Are there obvious language issues or emojis/symbols?

# Count of reviews
total_reviews = dataset['text'].shape[0]

# reviews containing any non-ASCII character (accents, other scripts, or emojis)
non_ascii_mask = dataset['text'].astype(str).str.contains(r'[^\x00-\x7F]', regex=True)
non_ascii_count = non_ascii_mask.sum()

# rough emoji detection (common emoji codepoint ranges)
emoji_pattern = r'[\U0001F300-\U0001F5FF\U0001F600-\U0001F64F\U0001F680-\U0001F6FF\U0001F700-\U0001F77F\U0001F900-\U0001F9FF\U0001FA70-\U0001FAFF]'
emoji_mask = dataset['text'].astype(str).str.contains(emoji_pattern, regex=True)
emoji_count = emoji_mask.sum()

print(f"Total reviews: {total_reviews}")
print(f"Reviews with non-ASCII characters: {non_ascii_count} ({non_ascii_count/total_reviews:.2%})")
print(f"Reviews with emojis (approx): {emoji_count} ({emoji_count/total_reviews:.2%})")

# show a few examples to inspect
print("\nExamples (non-ASCII / emoji):")
display(dataset.loc[non_ascii_mask | emoji_mask, ['title', 'stars', 'text']].head(5))
```
